```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on="title")
```

```python
movies.head(1)
```

```python
movies.head(1).columns
```

```python
movies = movies[['movie_id','title','overview','status','genres','vote_average','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies["genres"][0]
```

```python
import ast
ast.literal_eval(movies["genres"][0])
```

```python
def fun(obj):
    ll = []
    for i in ast.literal_eval(obj):
        ll.append(i['name'])
    return ll
```

```python
movies['genres'] = movies['genres'].apply(fun)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(fun)
```

```python
movies
```

```python
def con3(obj):
    l = []
    n = 0
    for i in ast.literal_eval(obj):
        if n!=3:
            l.append(i['name'])
            n = n+1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(con3)
```

```python
movies['crew'][0]
```

```python
def dir_name(obj):
    ll = []
    for ss in ast.literal_eval(obj):
        if(ss['job']=='Director'):
            ll.append(ss['name'])
            break
    return ll
```

```python
movies['crew'] = movies['crew'].apply(dir_name)
```

```python
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head(1)
```

```python
movies['overview'] = movies["overview"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['genres'] = movies["genres"].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies["keywords"].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies["cast"].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies["crew"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['Tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df  = movies[['movie_id','title','status','vote_average','Tags']]
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x : " ".join(x))
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    t = []
    for i in text.split():
        t.append(ps.stem(i))
    return " ".join(t)
```

```python
new_df["Tags"] = new_df["Tags"].apply(stem)
```

```python
new_df["status"] = new_df["status"].apply(stem)
```

```python
new_df
```

```python
new_df["Tags"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
cv
```

```python
model = cv.fit_transform(new_df['Tags'])
```

```python
model.shape
```

```python
np.unique(model.toarray()[0])
```

```python
vector = model.toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new_df
```

```python
def recommended(movie):
    try:
        movie_index = new_df[new_df['title'] == movie].index[0]
        dis = similarity[movie_index]
        movies_list = sorted(list(enumerate(dis)),reverse=True, key=lambda x:x[1])[1:6]
        for i in movies_list:
            print(new_df.iloc[i[0]].status)
            print(new_df.iloc[i[0]].vote_average)
            print(new_df.iloc[i[0]].title)
    except:
        print("The movie you want is not in the list")
```

```python
name = input("Enter the movie name")
recommended(name)
```
